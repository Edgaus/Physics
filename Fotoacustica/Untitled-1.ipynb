```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams.update({'font.size': 14})
font = {'fontname':'Microsoft YaHei','size'   : 22}

plt.title('PA signal',**font)
plt.xlabel('Frequency (Hz)', **font)
plt.ylabel('Amplitude (A.U.)', **font)


plt.rcParams.update({'font.size': 10})

m_207 = pd.read_excel('m_207.xlsx')
plt.plot( m_207['Frec_fit'], m_207['R_fit'], color='b', linewidth=2.0, label='Signal'  )
plt.scatter( m_207['Fret'], m_207['Rval'] ,color='r', marker="^", label='Fit'   )
plt.text( 100, 0.05, '$\\alpha = 6.37 \\pm 1.33$', fontsize = 22)
plt.legend()
```
